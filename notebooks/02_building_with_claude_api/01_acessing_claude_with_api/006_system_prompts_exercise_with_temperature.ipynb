```python
# Load env variables
from dotenv import load_dotenv

load_dotenv("/workspace/.env")
```

```python
#Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    
    if system:
        params["system"] = system
        
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# Start with an empty message list
messages = []

#Use a while true loop to have a multi-turn conversation forever
while True:
    #Get user input
    user_input = input("> ")
    print(">", user_input)

    # Add the user input to the conversation history
    add_user_message(messages, user_input)

    # Call claude with the chat function
    # With system prompt
    system = """
    You are a python expert engineer that helps people learn programming,
    and you always provide code examples in your answers,
    but this code are very direct, very concise and don't include any comments or explanations,
    because the user dont want that, they just want the code, and let's use the Keep It Simple Stupid (KISS) principle in our answers.
    """
    answer = chat(messages, system=system, temperature=0.7)

    # Add Claude's response to the conversation history
    add_assistant_message(messages, answer)

    # Print Claude's response
    print("---")
    print(answer)
    print("---")
```

```python
# Claude suggests this code below to find the Fibonacci sequence until 100, you can try it out in the conversation above.
def fibonacci_until_100():
    a, b = 0, 1
    while a < 100:
        print(a)
        a, b = b, a + b

fibonacci_until_100()
```
